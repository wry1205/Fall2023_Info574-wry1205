```python
import pandas as pd
```

```python
# netflix_titles
df1 = pd.read_csv('netflix_titles.csv')

print(df1)
```

```python
# netflix_titles_2
df2 = pd.read_csv('netflix_titles_2.csv')

print(df2)
```

```python
# as we want to figure out what characteristic may influence the popularity, show_id and title are meaningless, so drop them to simplify
df1 = df1.drop(columns = ['show_id', 'title'])
df2 = df2.drop(columns = ['show_id', 'title'])
```

```python
frames = [df1, df2]
raw_result = pd.concat(frames)
result = pd.concat(frames).drop_duplicates()

print(raw_result)
print(result)
```
